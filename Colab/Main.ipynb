```
from imageai.Detection import VideoObjectDetection
import os

def videoDetection(in_filename:str, out_filename:str, model_filename:str, **types):
  execution_path = os.getcwd()
  detector = VideoObjectDetection()
  detector.setModelTypeAsYOLOv3()
  detector.setModelPath( os.path.join(execution_path , model_filename))
  detector.loadModel(detection_speed="fast")
  custom_objects = detector.CustomObjects(**types) #Выбор отображаемых объектов

  def FrameDoing(frame_number, output_array, output_count):
    print("Frame: ", frame_number)
    print("Objects: ", output_count)

  video_path = detector.detectCustomObjectsFromVideo(
    custom_objects=custom_objects,
	  input_file_path=os.path.join(execution_path, in_filename),
	  output_file_path=os.path.join(execution_path, out_filename),
	  frames_per_second=20,
    per_frame_function=FrameDoing,
    minimum_percentage_probability=40, #Минимальный процент для отображения
	  log_progress=True
  )
  print(video_path)
```

```
videoDetection("VID.mp4", "OUT", "yolo.h5", person=True)
```
